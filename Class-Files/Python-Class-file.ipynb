    <a class="anchor-link" href="https://www.kaggle.com/code/shwetankmaurya/python-college" style="text-decoration:none; margin-left:10px;">¶</a>
</h1>

<div style="color: green; font-family: monospace; white-space: pre-wrap;">
1. ASCII (American Standard Code for Information Interchange) was published in the early 1960s.  
2. It uses 7 bits to represent characters, allowing 128 unique symbols (0 to 127).  
3. Examples: Lowercase 'a' to 'z' are from 97 to 122, and uppercase 'A' to 'Z' are from 65 to 90.  
4. ASCII is limited to English characters and basic symbols.  
5. To support characters from other languages, the Unicode standard was introduced.  
6. Python 3 uses Unicode (UTF-8) by default for strings.  
7. Unicode defines over 100,000 characters covering most of the world's writing systems.  
8. In Python 3, strings are Unicode by default, so the prefix <code>u</code> (like <code>u"hello"</code>) is optional.
</div>

9. <code>\'</code> → Single quote                             ---> ASCII: 39  
10. <code>\"</code> → Double quote                             ---> ASCII: 34  
11. <code>\0</code> → Null character (string terminator)       ---> ASCII: 0

<h1 id="basics" style="font-family:verdana; text-align:center;">

    <a class="anchor-link" href="https://www.kaggle.com/code/shwetankmaurya/python-college" style="text-decoration:none; margin-left:10px;">¶</a>
</h1>

<div style="color: pink; font-family: monospace; white-space: pre-wrap;">
<b>SPLIT METHOD</b>  
- Used to split a string into a list using a delimiter (default: whitespace).  
- Handles multiple consecutive spaces by default.  
- Syntax: <code>string.split([separator], [maxsplit])</code>  
- Example: <code>"Hello   world".split()</code> → <code>["Hello", "world"]</code>
<b>JOIN METHOD</b>  
- Used to join a list of strings into a single string using a specified separator.  
- Syntax: <code>"separator".join(iterable)</code>  
- Example: <code>"-".join(["2025", "10", "03"])</code> → <code>"2025-10-03"</code>  
- You can join without spaces, or use any delimiter like comma, space, dash, etc.
<b>Applications:</b>  
1. Useful in processing CSV data, text parsing, and formatting strings.  
2. Split long paragraphs into words/sentences and join them after processing.  
3. Clean up inconsistent spacing in raw text.
</div>

<div style="color: pink; font-family: monospace; white-space: pre-wrap;">
<b>FIND METHOD</b>  
- <b>Purpose:</b> Returns the index of the first occurrence of a specified substring.  
- <b>If Not Found:</b> Returns <code>-1</code> if the substring is not present.  
- <b>Syntax:</b> <code>string.find(substring)</code>  
- <b>Behavior:</b> Only returns the index of the <b>first match</b>.
<b>Example:</b>  
<code>"python is awesome".find("is")</code> → <code>7</code>  
<code>"python is awesome".find("java")</code> → <code>-1</code>
<b>Applications:</b>  
1. Useful in searching for keywords or patterns in text.  
2. Can be used in basic string validation or filtering logic.  
3. Acts as a safer alternative to <code>index()</code>, which raises an error if not found.

- Used to represent truth values in conditions and logical operations.

<code>is_greater = (7 &gt;= 7)  # True</code>

```python
# '==' checks if values are equal
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True because values are equal
print(a is b)  # False because they are different objects in memory

# 'is' checks if both variables point to the same object
c = a
print(a is c)  # True because both point to the same object
```

```python
# Define some sample conditions
condition1 = True
condition2 = False
condition3 = True

# Without all()
if condition1 and condition2 and condition3:
    print("All conditions met (without all())")
else:
    print("Not all conditions met (without all())")

# Using all()
conditions = [condition1, condition2, condition3]
if all(conditions):
    print("All conditions met (with all())")
else:
    print("Not all conditions met (with all())")

# Using any()
if condition1 or condition2 or condition3:
    print("At least one condition met (without any())")

if any(conditions):
    print("At least one condition met (with any())")
```

```python
# Example boolean conditions
has_id_card = True
paid_fees = True
submitted_assignment = False

# Check if student is eligible to sit in exam (all conditions must be True)
eligible = all([has_id_card, paid_fees, submitted_assignment])
print("Eligible to sit in exam:", eligible)  # False because submitted_assignment is False

# Check if student qualifies for participation certificate (any condition True)
participation = any([has_id_card, paid_fees, submitted_assignment])
print("Qualifies for participation certificate:", participation)  # True
```

<b>LIST</b>  
- A built-in sequence data type in Python used to store multiple items.  
- Defined using square brackets <code>[]</code> and items are separated by commas.  
- Lists are flexible, dynamic, and versatile.

<b>Properties of List:</b>  

1. <b>Ordered:</b> Elements in a list retain their order.  
   <code>numbers = [10, 20, 30]</code> → Order is preserved when accessed.  

2. <b>Allows Duplicates:</b> Same value can appear multiple times.  
   <code>data = [1, 1, 2, 3]</code> → No restriction on duplicates.  

3. <b>Mutable:</b> You can change, add, or delete elements.  
   <code>items = [5, 6, 7]</code>  
   <code>items[1] = 99</code> → <code>[5, 99, 7]</code>  

4. <b>Heterogeneous Elements:</b> List can store different data types.  
   <code>mixed = [1, "hello", 3.14, True]</code>  

5. <b>Dynamic Size:</b> Lists grow or shrink automatically.  
   <code>my_list.append(4)</code> → Adds item  
   <code>my_list.pop()</code> → Removes last item  

6. <b>Indexing Supported:</b> Access elements using positive or negative indexes.  
   <code>fruits = ["apple", "banana", "cherry"]</code>  
   <code>fruits[0]</code> → <code>"apple"</code>, <code>fruits[-1]</code> → <code>"cherry"</code>  

7. <b>Slicing Supported:</b> Extract parts of the list using slice notation.  
   <code>nums = [10, 20, 30, 40]</code>  
   <code>nums[1:3]</code> → <code>[20, 30]</code>  

8. <b>Nested Lists:</b> A list can contain other lists (2D or more).  
   <code>matrix = [[1, 2], [3, 4]]</code> → List inside a list  

9. <b>Iterable:</b> You can loop through a list using a loop.  
   <code>for x in [1, 2, 3]: print(x)</code>  

10. <b>Multiple Built-in Methods:</b> Lists have many useful methods.  
    <code>append()</code>, <code>extend()</code>, <code>insert()</code>, <code>remove()</code>, <code>sort()</code>, <code>reverse()</code>, etc.

<b>SORTING IN PYTHON: KEY POINTS</b>

- <b><code>sort()</code> vs <code>sorted()</code></b>  
  - <code>list.sort()</code> modifies the list in-place and returns <code>None</code>.  
  - <code>sorted(iterable)</code> returns a new sorted list and does not change the original.  

- <b>Default Behavior</b>  
  - Sorts in <b>ascending order</b>.  
  - For strings, sorting is <b>case-sensitive</b> (e.g., <code>'Zebra' &lt; 'apple'</code>).  

- <b>Reverse Sorting</b>  
  - Use <code>reverse=True</code> to sort in descending order.  
    <code>numbers.sort(reverse=True)</code>  

- <b>Custom Key Sorting</b>  
  - Use the <code>key</code> argument to customize how elements are compared.  
    <code>names.sort(key=str.lower)</code>  (case-insensitive sort)  

- <b>Stable Sort</b>  
  - Python's sort is <b>stable</b>: equal elements maintain their relative order.  

- <b>Sorting with Lambda</b>  
  - Use <code>lambda</code> functions for custom sorting logic:  
    <code>people.sort(key=lambda x: x[1])</code>  (sort by second element)  

- <b>Sorting Lists of Tuples/Objects</b>  
  - Sort by specific index or attribute:  
    <code>students = [("Alice", 90), ("Bob", 85)]</code>  
    <code>sorted(students, key=lambda x: x[1])</code>  (sort by score)  

- <b>Don't Sort Mixed Data Types</b>  
  - Sorting mixed types like <code>int</code> and <code>str</code> raises <code>TypeError</code> in Python 3.  
    <code>sorted([1, "apple"])</code>  (raises error)  

<b>Example:</b>  
<code>words = ["banana", "Apple", "cherry"]</code>  
<code>words.sort()</code>  # Case-sensitive sort  
<code>words.sort(key=str.lower)</code>  # Case-insensitive sort

<div style="color: pink; font-family: monospace; white-space: pre-wrap; background-color:#1e1e2f; padding:15px; border-radius:5px;">
A function is a named block of reusable code designed to perform a specific task.

Functions help in organizing programs into logical, manageable sections.

They can accept inputs (arguments) and optionally return outputs.

<span style="color:#8be9fd; font-weight:bold;">Definition:</span>

Defined using the <code>def</code> keyword followed by the function name.

Functions can have parameters inside parentheses; these act as placeholders for values.

A function may return a value using the <code>return</code> statement.

<span style="color:#50fa7b; font-weight:bold;">Why use functions?</span>

Avoid repeating code.

Increase readability and maintainability.

Separate logic into clear parts.

Make debugging easier.

<span style="color:#ffb86c; font-weight:bold;">ARGUMENTS (Inputs to a Function)</span>

Arguments are values you pass into a function so it can work with specific data.

Types of arguments:

<b>Positional arguments</b>: Passed in order; position matters.

<b>Keyword arguments</b>: Passed by name, making order irrelevant.

<b>Default arguments</b>: Parameters with preset values used when no value is provided.

<b>Example (very basic):</b>
<code>def greet(name="Guest"):</code>
This means the function has one parameter with a default value.

<span style="color:#bd93f9; font-weight:bold;">*ARGS (Variable-Length Positional Arguments)</span>

<code>*args</code> allows a function to accept any number of positional arguments.

The function receives them as a tuple.

Useful when you don’t know how many values will be passed in.

<b>Notes:</b>

The name "args" is a convention, but any name with * works.

Order matters since they are positional.

Commonly used for tasks where inputs vary in quantity.

<b>Basic structure:</b>
<code>def example(*args):</code>
Inside the function, <code>args</code> behaves like a tuple containing all passed values.

<span style="color:#ff5555; font-weight:bold;">**KWARGS (Variable-Length Keyword Arguments)</span>

<code>**kwargs</code> allows a function to accept any number of keyword arguments.

The function receives them as a dictionary (key-value pairs).

Useful when passing flexible or descriptive data.

<b>Notes:</b>

“kwargs” is the standard name, but any name with ** works.

Order does not matter because arguments are named.

Ideal for configuration-like parameters.

<b>Basic structure:</b>
<code>def example(**kwargs):</code>
Inside the function, <code>kwargs</code> behaves like a dictionary.

<span style="color:#f1fa8c; font-weight:bold;">USING *ARGS AND **KWARGS TOGETHER</span>

Both can be used in the same function definition.

Order must be:

regular parameters
*args
keyword-only parameters (optional)
**kwargs

Allows maximum flexibility: any number of unnamed + named inputs.

<span style="color:#69ff94; font-weight:bold;">BENEFITS OF *ARGS AND **KWARGS</span>

Allow flexible function calls.

Reduce the need to define many parameters.

Useful for writing general-purpose or highly reusable functions.

Make functions more adaptable as requirements grow.

<span style="color:#ff92df; font-weight:bold;">DOCSTRINGS</span>

Functions can include a docstring (a string under the definition) describing purpose and usage.

Helps with documentation and readability.

</div>

```python
def arms(a: int) -> bool:
    digits = str(a)
    power = len(digits)
    total = sum(int(d)**power for d in digits)
    return total == a

g=arms(153)
print(g)
```

```python
def armss(a: int) -> bool:
    total = 0
    aa = a
    n = len(str(a))
    while a > 0:
        digit = a % 10
        total += digit ** n
        a = a // 10
    return total == aa

armss(34)
```

Question : write a menu based program to create a file, append contents in a file, replace contents in file, delete a file and search a word in a file.
</p>
</div>


```python
import os

def create_file(filename):
    with open(filename, "w") as f:
        print("Enter data (end with a single #):")
        while True:
            line = input()
            if line == "#":
                break
            f.write(line + "\n")
    print("File created successfully.")
def append_file(filename):
    if not os.path.exists(filename):
        print("File does not exist!")
        return
    with open(filename, "a") as f:
        print("Enter data to append (end with #):")
        while True:
            line = input()
            if line == "#":
                break
            f.write(line + "\n")
    print("Data appended successfully.")
def search_word(filename, word):
    if not os.path.exists(filename):
        print("File not found!")
        return
    found = False
    with open(filename, "r") as f:
        for line_no, line in enumerate(f, start=1):
            if word in line:
                print(f"Found in line {line_no}: {line.strip()}")
                found = True
    if not found:
        print("Word not found.")
def replace_word(filename, old, new):
    if not os.path.exists(filename):
        print("File not found!")
        return
    with open(filename, "r") as f:
        data = f.read()
    data = data.replace(old, new)
    with open(filename, "w") as f:
        f.write(data)
    print("Word replaced successfully.")

def delete_file(filename):
    if os.path.exists(filename):
        os.remove(filename)
        print("File deleted successfully.")
    else:
        print("File not found.")
while True:
    print("\n===== FILE HANDLING MENU =====")
    print("1. Create File")
    print("2. Append Data")
    print("3. Search Word")
    print("4. Replace Word")
    print("5. Delete File")
    print("6. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        filename = input("Enter filename: ")
        create_file(filename)
    elif choice == "2":
        filename = input("Enter filename: ")
        append_file(filename)
    elif choice == "3":
        filename = input("Enter filename: ")
        word = input("Enter word to search: ")
        search_word(filename, word)

    elif choice == "4":
        filename = input("Enter filename: ")
        old = input("Enter word to replace: ")
        new = input("Enter new word: ")
        replace_word(filename, old, new)
    elif choice == "5":
        filename = input("Enter filename to delete: ")
        delete_file(filename)
    elif choice == "6":
        print("Exiting...")
        break
    else:
        print("Invalid choice! Try again.")
```
