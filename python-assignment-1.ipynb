<p style="padding: 10px;
              color:white;">
            Question-4.  Create a function<code>heart_rate_stats(max_bpm, avg_bpm, resting_bpm)</code> that returns:<br> 
‚óè Heart Rate Reserve = max_bpm - resting_bpm <br>
‚óè Average Difference = avg_bpm - resting_bpm <br>
Print both returned values separately.            
</p>
</div>

```python
def heart_rate_stats(max_bpm, avg_bpm, resting_bpm):  # defining the heart rate stats function 
    heart_rate_reserve = max_bpm - resting_bpm    # Here basic calculations to get the rate of the reserve and the average difference
    average_difference = avg_bpm - resting_bpm
    return heart_rate_reserve, average_difference    # returning the values

# Taking the input
max_bpm=int(input("Enter the number of maximum times heart beats in a minute:"))
avg_bpm=int(input("Enter the average number of times heart beats in a minute:"))
resting_bpm=int(input("Enter the number of times heart beats when resting:"))
 # passing the values into the function
reserve,difference=heart_rate_stats(max_bpm,avg_bpm,resting_bpm)
# printing the values
print('-'*40)
print("Heart Rate Reserve:",reserve)
print("Average Difference:",difference)
print('-'*40)
```

```python
def sum_n(n): # defining a function which recurssively calculates the sum
    if(n==1): # base case is n==1 then return 1
        return 1
    return sum_n(n-1)+n # recursive call
n=int(input("Enter the number:")) # taking the input 
print(f"Sum of the first {n} natural number:{sum_n(n)}") # printing the output
```

```python
def factorial(n): # here is the function which prints the factorial of the number
    if(n==1): # base case if n=1 then returns 1
        return 1
    return (n)*factorial(n-1) # otherwise return the n multiplys with one less than n
    
n=int(input("Enter the number:")) # taking the input
print(f"Factorial of {n} is {factorial(n)}.") #  printing factorial of the number 
```
